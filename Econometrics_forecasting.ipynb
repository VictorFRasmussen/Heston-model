```python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.stats import norm
import matplotlib.pyplot as plt
```

# Notebook for creating econometric forecasts
The purpose of this notebook is to test different econometric models, for the purpose of financial forecasting, with the main focus on stock prices

```python
msft = yf.Ticker("MSFT")


# get historical market data
hist = msft.history(period="1mo")

print(hist)
```

```python
msft_price = hist['Close']

plt.figure(figsize=(10, 6))
plt.plot(msft_price, linestyle='-', color='b', label='MSFT price')
plt.ylabel('price USD')
plt.title(f'MSFT price last year')
plt.grid(True)
plt.legend()
plt.show()
```

```python
log_msft_price = np.log(msft_price)
plt.figure(figsize=(10, 6))
plt.plot(log_msft_price, linestyle='-', color='b', label='MSFT price')
plt.ylabel('price USD')
plt.title(f'MSFT price last year')
plt.grid(True)
plt.legend()
plt.show()
```
